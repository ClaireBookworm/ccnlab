# Simulation and Evaluation

```python
import random

# Example of what a user's custom model may look like. The benchmark is agnostic
# to how the model is internally structured.
class RandomModel:
    """Produces response with probability that changes linearly with each US."""
    def __init__(self, start=0.2, delta=0.1, min_prob=0.1, max_prob=0.9):
        self.prob = start
        self.start = start
        self.delta = delta
        self.min_prob = min_prob
        self.max_prob = max_prob
        
    def reset(self):
        self.prob = start
    
    def act(self, cs, ctx, us):
        if us > 0:
            self.prob = max(min(self.prob + self.delta, self.max_prob), self.min_prob)
            return 1
        if len(cs) > 0:
            return random.choices([1, 0], weights=[self.prob, 1-self.prob])[0]
        return 0
```

```python
import ccnlab.benchmarks.classical as classical
import ccnlab.evaluation as evaluation

# Select experiments to run, filtering by name using glob syntax.
for exp in classical.registry('*'):
    
    # Print experiment information.
    display_heading(exp.name, level=2)
    print(exp.meta['citation'])
    print(exp.schedule())
    
    # Experiments often have multiple groups, each shown different stimuli.
    for g, group in exp.stimuli.items():
        
        # Users are free to decide how to initialize their models and how many
        # instances to allocate per group.
        for instance in range(1):
            model = RandomModel()
            
            # Each group is shown a sequence of trials. At each timestep in a
            # trial, the model input consists of the conditioned stimuli (cs),
            # context (ctx), and unconditioned stimulus (us); the model output
            # should be a binary response value.
            for i, trial in enumerate(group):
                for t, timestep in enumerate(trial):
                    cs, ctx, us = timestep
                    res = model.act(cs, ctx, us)
                    exp.data[g][i][t]['response'].append(res)

    # Simulation results can be compared to empirical results from published
    # work, via evaluation metrics (quantitative) and plots (qualitative).
    empirical = exp.empirical_results
    simulated = exp.simulated_results()
    print('correlation:', evaluation.correlation(empirical, simulated))
    display_side_by_side(empirical, simulated)
    exp.plot(show='both')
    plt.show()
```

# Specification

```python
import ccnlab.benchmarks.classical.core as cc

class TestExperiment(cc.ClassicalConditioningExperiment):
    def __init__(self):
        super().__init__({
            'control': cc.seq(cc.trial('AB+'), repeat=6),
            'special': cc.seq(cc.trial('AB+'), cc.trial('A+'), repeat=6),
        })
        self.empirical_results = pd.melt(pd.DataFrame(
            columns=['group', 'trial', 'AB'],
            data=[
                ['control', 1, 0.2],
                ['control', 2, 0.3],
                ['control', 3, 0.4],
                ['control', 4, 0.5],
                ['control', 5, 0.6],
                ['control', 6, 0.7],
                ['special', 1, 0.2],
                ['special', 2, 0.3],
                ['special', 3, 0.4],
                ['special', 4, 0.5],
                ['special', 5, 0.6],
                ['special', 6, 0.7],
            ]
        ), id_vars=['group', 'trial'])
        self.meta = dict(ylabel='hi')
        self.plots = [
            lambda df, ax, **info: cc.plot_lines(df, ax=ax),
        ]

    def simulated_results(self):
        return pd.melt(self.dataframe(lambda x: {
            'AB': cc.conditioned_response(x['trial'], x['responses'], ['A', 'B']),
        }), id_vars=['group', 'trial']).groupby(['group', 'trial', 'variable']).mean().reset_index()
```
